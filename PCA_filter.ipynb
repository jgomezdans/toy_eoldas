```python
xleafn = 1.
cab = 20 + np.random.rand(100)*80
car = 15.
cb = 0.1
cw = 0.001
cm = 0.0001
lai = np.random.rand ( 100 )*6+2
lidfa = -0.3
lidfb = 0.3
hspot = 0.02
tts = 29
tto = 0
phi = 156
rhos = 0.5
rho = [ prosail.run_prosail ( xleafn, cab[i], car, cb, cw, cm, lai[i], \
      lidfa, lidfb, rhos, hspot, tts, tto, phi ) for i in xrange(cab.shape[0]) ]
```
